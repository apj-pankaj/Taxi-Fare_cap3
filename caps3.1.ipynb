```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df1=pd.read_csv("D:/DS ML&AI/Capstone3/env/Scripts/cleaned.csv")
df1
```

```python
df1.select_dtypes(include=['object']).columns
```

```python
cat_columns=['store_and_fwd_flag','pickup_day','AM/PM']
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
#OneHotEncoder for categirical columns
ohe = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

ohe_col = ohe.fit_transform(df1[cat_columns])

ohe_df = pd.DataFrame(ohe_col, columns=ohe.get_feature_names_out(cat_columns))
ohe_df = ohe_df.astype(int)

df2=pd.concat([df1.drop(columns=cat_columns), ohe_df], axis=1)
```

```python
df2['tpep_pickup_datetime'] = pd.to_datetime(df2['tpep_pickup_datetime'])
df2['tpep_dropoff_datetime'] = pd.to_datetime(df2['tpep_dropoff_datetime'])
```

```python
df2['tpep_pickup_datetime_utc'] = df2['tpep_pickup_datetime'].dt.tz_localize('UTC')
df2['tpep_pickup_datetime_edt'] = df2['tpep_pickup_datetime_utc'].dt.tz_convert('US/Eastern')
df2['tpep_dropoff_datetime_utc'] = df2['tpep_dropoff_datetime'].dt.tz_localize('UTC')
df2['tpep_dropoff_datetime_edt'] = df2['tpep_dropoff_datetime_utc'].dt.tz_convert('US/Eastern')
```

```python
df2=df2.drop(columns=['tpep_pickup_datetime', 'tpep_dropoff_datetime', 'tpep_pickup_datetime_utc' ,'tpep_pickup_datetime_edt', 'tpep_dropoff_datetime_utc', 'tpep_dropoff_datetime_edt']) 
```

```python
#feature selection with threshhold variance
from sklearn.feature_selection import VarianceThreshold
```

```python
X = df2.drop(columns=['total_amount'])
y = df2['total_amount']

vt=VarianceThreshold(threshold=0.5)
X_new=vt.fit_transform(X)


sel_col=vt.get_feature_names_out()

X_new=pd.DataFrame(X_new,columns=sel_col)

#X_new


feature_importance_df = pd.DataFrame({
    'Feature': X.columns,
    'Importance': vt.variances_
    })

# Sort by importance in descending order
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)


feature_importance_df
```

```python
feature_importance_df =feature_importance_df.reset_index(drop=True) 
feature_importance_df
```
